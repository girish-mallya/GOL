```python
import numpy as np
import random as rd
cols=8


def initmat():
    x = [[0 for i in range(8)] for i in range(8)]
    #print(len(x))
    return(x)

def randomstart(x):
    #print(x)
       
    for z in range(len(x)):
        i=rd.randint(0,7)
        #print(i)
        j=rd.randint(0,7)
       # print(j)
        x[i][j]=1
        
def printmat(x):
    for z in range(len(x)):
        print(x[z])
        
def getsumofneighbors(matrix, i, j):
    tempsum=0
    if(i-1>0):
        tempsum=tempsum+matrix[i-1][j]
    if(i+1<cols):
        tempsum=tempsum+matrix[i+1][j]
    if(j+1<cols):
        tempsum=tempsum+matrix[i][j+1]
        if(i+1<cols):
            tempsum=tempsum+matrix[i+1][j+1]
        if(i-1>0):
            tempsum=tempsum+matrix[i-1][j+1]
    if(j-1>0):
        tempsum=tempsum+matrix[i][j-1]
        if(i+1<cols):
            tempsum=tempsum+matrix[i+1][j-1]
        if(i-1>0):
            tempsum=tempsum+matrix[i-1][j-1]
    return(tempsum)
        
        
            
            

def rule1(x):
    #Any live cell with fewer than two live neighbors dies, as if by under population.
    for i in range(len(x)):
        count=0
        for j in range(len(x[i])):
            tsum=getsumofneighbors(x,i,j)
            #print(tsum)
            if tsum<1:
                x[i][j]=0
            
            
def rule2(x):
    #Any live cell with two or three live neighbors lives on to the next generation.
     for i in range(len(x)):
        #count=0
        for j in range(len(x[i])):
            if x[i][j]==1:
                tsum=getsumofneighbors(x,i,j)
                
                if tsum==2 or 3:
                    x[i][j]=1
def rule3(x):
    #Any live cell with more than three live neighbors dies, as if by overpopulation.
     for i in range(len(x)):
        #count=0
        for j in range(len(x[i])):
            if x[i][j]==1:
                tsum=getsumofneighbors(x,i,j)
                
                if tsum>3:
                    x[i][j]=0


def rule4(x):
    #Any dead cell with exactly three live neighbors becomes a live cell, as if by reproduction.
     for i in range(len(x)):
        #count=0
        for j in range(len(x[i])):
            if x[i][j]==0:
                tsum=getsumofneighbors(x,i,j)
                
                if tsum==3:
                    x[i][j]=1

                        
def main():
    x=[[]]
    x=initmat()
    randomstart(x)
    printmat(x)
    rule1(x)
    print("after rule1")
    printmat(x)
    rule2(x)
    print("after rule2")
    printmat(x)
    rule3(x)
    print("after rule3")
    printmat(x)
    rule4(x)
    print("after rule4")
    printmat(x)
```

```python
main()
```
